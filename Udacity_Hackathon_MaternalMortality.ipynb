```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
%matplotlib inline
```

```python
#Reading MaternalMortality.csv into a dataframe with index as country
df_maternalmortality = pd.read_csv("MaternalreproductiveHealth\MaternalMortality.csv",index_col='Country')
```

```python
#Printing the dataframe
df_maternalmortality
```

```python
#Checking Datatypes of the dataframe columns
df_maternalmortality.dtypes
```

```python
#Changing the Year Datatype to string
df_maternalmortality['Year']=df_maternalmortality['Year'].astype('str')
df_maternalmortality['Year']=df_maternalmortality['Year'].str.strip()
df_maternalmortality
```

```python
#Removing the ranges from The Maternal mortality ratio
df_maternalmortality['Maternal mortality ratio (per 100 000 live births)'] = df_maternalmortality['Maternal mortality ratio (per 100 000 live births)'].astype(str)
df_maternalmortality['Maternal mortality ratio (per 100 000 live births)']=df_maternalmortality['Maternal mortality ratio (per 100 000 live births)'].apply(lambda x: x.split('[')[0])
```

```python
#Changing Maternal Mortality Ratio to float for analysis
df_maternalmortality['Maternal mortality ratio (per 100 000 live births)'] = df_maternalmortality['Maternal mortality ratio (per 100 000 live births)'].str.replace(" ","")
df_maternalmortality['Maternal mortality ratio (per 100 000 live births)'] = df_maternalmortality['Maternal mortality ratio (per 100 000 live births)'].astype(float)
```

```python
#Filling null values with zeros
df_maternalmortality.fillna(0,inplace=True)
df_maternalmortality
```

```python
#Since there are alot of null values, for analysis purposes I am not taking years into account 
#and taking average of the years the data is available
df = df_maternalmortality.mean(level=0)
df
```

```python
df.plot(y='Maternal mortality ratio (per 100 000 live births)',kind='bar',figsize=(30,10))
```

```python
df.plot(y='Births attended by skilled health personnel (%)',kind='bar',figsize=(30,10))
```

```python
df.sort_values('Maternal mortality ratio (per 100 000 live births)', ascending=False).head(10)
```

```python
#Making regression plot to analyze the relationship between Maternal Mortality & births attaended by skilled physician
sns.regplot(x='Births attended by skilled health personnel (%)',y='Maternal mortality ratio (per 100 000 live births)',data=df)
plt.show()
```
